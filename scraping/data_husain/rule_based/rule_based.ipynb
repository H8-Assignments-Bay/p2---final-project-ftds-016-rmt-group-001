```python
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

def categorize_comment(row, symbols):
  comment = row['comment']
  
  # Tokenisasi komentar menjadi kata-kata
  words = nltk.word_tokenize(comment)
  
  # Menggunakan Porter Stemmer untuk mengubah kata-kata menjadi bentuk dasar
  stemmer = nltk.PorterStemmer()
  stemmed_words = [stemmer.stem(word) for word in words]
  
  # Menggunakan part-of-speech tagging untuk mengidentifikasi jenis kata
  pos_tags = nltk.pos_tag(words)
  
  # Cek apakah kata dasar dari setiap symbol ada di komentar
  for symbol in symbols:
    if stemmer.stem(symbol) in stemmed_words:
      return symbol
  
  # Cek apakah kata dasar dari setiap symbol ada di komentar sebagai subjek (noun) atau objek (noun)
  for symbol in symbols:
    for word, pos in pos_tags:
      if stemmer.stem(symbol) == stemmer.stem(word) and (pos == 'NN' or pos == 'NNP'):
        return symbol
  
  return "Other"

# Contoh penggunaan fungsi
symbols = ['FISH', 'JARR', 'FB']
df['category'] = df.apply(lambda row: categorize_comment(row, symbols), axis=1)
df
```

```python
import re
import nltk
nltk.download('punkt')

def is_subject_equal_to_symbol(row):
    """
    subject is the stock symbol discussed in the comment.
    symbol is the stock where the comment came from.
    
    Return True if subject == symbol.
    Return False if subject != symbol.
    """
    comment = row['comment']

    # Tokenisasi kolom comment menjadi kata-kata
    words = nltk.word_tokenize(comment)

    # Buat list kosong untuk menyimpan kata-kata huruf besar yang ditemukan di dalam kolom comment
    uppercase_words = []

    # Periksa apakah terdapat kata huruf besar semua di dalam kolom comment
    for word in words:
        if word.isupper():
            uppercase_words.append(word)

    # Jika terdapat 4 kata huruf besar semua dan kata selanjutnya adalah ',' beserta 4 kata huruf besar lainnya, maka return 'accept'
    if len(uppercase_words) == 5 and uppercase_words[-1] == ',':
        return "Accept"

    # Mencari 4 kata huruf besar semua dan diawali simbol $ beserta tanda
    pattern = r"\$[A-Z]{4}(\s|,| dan)\s\$[A-Z]{4}"
    if re.search(pattern, comment):
        return "accept"

    # Tokenisasi komentar menjadi kata-kata
    words = nltk.word_tokenize(comment)

    # Menggunakan Porter Stemmer untuk mengubah kata-kata menjadi bentuk dasar
    stemmer = nltk.PorterStemmer()
    stemmed_words = [stemmer.stem(word) for word in words]

    # Cek apakah komentar merupakan pertanyaan saham
    if 'apa' in stemmed_words or \
        'kenapa' in stemmed_words or \
        'bagaimana' in stemmed_words or \
        'gimana' in stemmed_words or \
        'kapan' in stemmed_words or \
        'dimana' in stemmed_words or \
        'siapa' in stemmed_words:
        return "Reject"

    # Cari subjek komentar
    subject = None
    for word in words:
        if word.isupper():
            subject = word
            break

    # Jika tidak ada subjek komentar, maka ACCEPT
    if subject is None:
        return "Accept"

    # Cari symbol pertama dari subjek komentar
    if stemmer.stem(row['symbol']) == stemmer.stem(subject):
        return 'Accept'
    else:
        return 'Reject'
```

```python
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
import pandas as pd
import re

def get_subjects(comment: str):
    subjects = re.findall(pattern="(?:\$?[A-Z]{4},?\s?){1,}(?:(?:dan)?(?:hingga)?\s?[A-Z]{4})?", string=comment)

    subjects = [re.sub("[\$\n,]|dan|hingga", "", item).split() for item in subjects]
    
    return subjects
  
def is_subject_equal_to_symbol(subjects: list[list[str]], symbol: str):
  
  for index, finds in enumerate(subjects):
    # check 1st subject
    if index == 0:
      # no subject
      # e.g "the stock is good"
      if len(finds) == 0: return 1
      # 1st subject length is 1
      # e.g "GOTO is bad", "ICBP is good"
      if len(finds) == 1:
        if finds[0] == "IHSG": continue
        elif finds[0] == symbol: return 1
        else: return 0
    
    # check 1st and 2nd subject with more than 1 subject
    # e.g 
    if len(finds) == 1:
      if finds[0] == symbol: return 1
    else:
      for find in finds:
        if find == symbol: return 1

  return 0
  
df = df[df['label_2'] == 0].copy()
df['subjects'] = df['comment'].apply(lambda comment: get_subjects(comment))
df[''] = df[['subjects', 'symbol']].apply(lambda row: is_subject_equal_to_symbol(row['subjects'], row['symbol']), axis=1)

with pd.option_context('display.max_colwidth', None, 'display.max_rows', None):
  display(df[['symbol', 'comment', 'label_1', 'subjects', 'subject']])
```
